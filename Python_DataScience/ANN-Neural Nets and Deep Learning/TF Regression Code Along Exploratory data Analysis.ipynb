Now we will scale our dataFrame and apply train test split

```python
x=df.drop('price',axis=1).values
y=df['price'].values
```

in above code '.values' has been used which will return the numpy array underneath the actual dataFrame or pandas core series 

we will do it to remove data types errors 

```python
x1=df['price']
type(x1)
```

In above result we see that without '.values' the type of 'x1' is pandas core series

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

We always do scalling post split to avoid data leakage from test set 

Check Below :--->>>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

In above query 'X_test' has only been transformed not 'fit' bcz we dont want to get prior information about our test 

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

Since There are 19 features in our dataFrame

we will keep 19 neurons in our Layers

and our activation layer will be rectified linear unit

```python
model=Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
```

```python
pd.DataFrame(model.history.history)
```

In above results we can see both loss on training data and as well as validation data(test data here)

and we can compare both loss using plots

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

In above plot we see that training loss and validation loss are same 
thus this is not the case of overfitting and we could continue our training 

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```
