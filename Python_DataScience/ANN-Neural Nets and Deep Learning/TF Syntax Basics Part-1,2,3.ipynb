```python
import numpy as np 
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_file_path="D:/DataScience_Udemy/TensorFlow_FILES/DATA/fake_reg.csv"
```

```python
df=pd.read_csv(input_file_path)
```

```python
df.head()
```

```python
sns.pairplot(df)
```

#### In above plot we can see that 'feature2' is highly colinear with 'price'

```python
from sklearn.model_selection import train_test_split
```

in below example we are making a pandas dataframe as numpy ndarray that's why we have put ".values" in the end

```python
X=df[['feature1','feature2']].values
```

```python
X
```

```python
y=df['price'].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print("\n")
print(X_test.shape)
```

#### Data will be scaled before training the model 

#### Thus model will be trained on scaled data

***Check Below:-->>***

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
print(X_train.min())
print(X_train.max())
```

```python
X_test.min()
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test.min()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
help(Sequential)
```

```python
model=Sequential()
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=250)
```

```python
model.history.history
```

```python
loss_df=pd.DataFrame(model.history.history)
print(loss_df.head())
print(loss_df.tail())
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions=model.predict(X_test)
```

```python
type(test_predictions)
```

```python
test_predictions=pd.Series(test_predictions.reshape(300,))
```

```python
test_predictions
```

```python
type(y_test)
```

```python
pred_df=pd.DataFrame(y_test,columns=['True Y test'])
```

```python
pred_df.head()
```

```python
pred_df=pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns=['True Y test','Model predictions Y']
```

```python
pred_df.head()
```

```python
sns.scatterplot(x='True Y test',y='Model predictions Y',data=pred_df)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(pred_df['True Y test'],pred_df['Model predictions Y'])
```

****Since in model loss function had been taken as 'mse' there also 'mse'=25....****

***Here also its coming as 25.231146 that's correct***

```python
mean_absolute_error(pred_df['True Y test'],pred_df['Model predictions Y'])
```

#### In above result mean_absolute_error is coming as 4.0339 which means predicted price will be '4.03397' up or down from original price. 

#### Below,In orginal dataframe  mean price is closer to 500 thus comparatively to mean price mean error is near about '1%' thus this model is can be considered a good model
