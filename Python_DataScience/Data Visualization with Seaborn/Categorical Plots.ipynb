```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

# ***working with barplot***

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

***In the above example of barplot*** 

***average of total_bill is showing for male and female***

*if we want to check standard deviation of category male and female*

*then we can use ***estimator*** argument and provide required functions*

*Check Below:--->>>*

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

*In the above example **standard deviation** *is calculated on total_bill column for both category male and female*

# Working with ***countplot***

***it gives count of a column*** 

***check Below :--->>***

```python
sns.countplot(x='sex',data=tips)
```

```python
tips.groupby(['sex']).count()
```

# working with "***boxplot***"

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

***Adding hue in boxplot to make it more fancy***

***Check Below :--->>>***

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
tips.groupby(['smoker','day'])['total_bill'].min()
```

```python
tips.head()
```

# Working with ***violinplot***

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# Working with "***stripplot***"

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

***In normal stripplot density of data points is not much visible***

***Check below:-->>***

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

***If jitter=False then it won't be much clear to see the difference in data points***

# Working with "***swarmplot***"

***if idea of violinplot and stripplot is combined it gives swarplot***

***and better visibility of datapoints***

***Check Below:--->>>***

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

# Disadvantages of swarmplot 

***Not fit for very large dataset***-takes alot time to arrange the datapoints

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='strip',hue='sex')
```
