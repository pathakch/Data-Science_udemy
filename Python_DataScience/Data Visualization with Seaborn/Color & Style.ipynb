# ***Removing Spines***

***Check Below:--->>>***

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

# In ***sns.despine()*** right and top spine removal is by default

***to remove bottom and left spine we need to give "bottom=True and left=True"**

***Check Below-->>>***

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

# Working with Scale & Context

***seaborn comes with set_context() to create plots and scale required***

***Check Below:--->>>***

```python
sns.set_context('poster',font_scale=0.5)
sns.countplot(x='sex',data=tips)
```

**Above example is for poster context if we want to paste it on a poster**

***Here fontsize also can be changed to different values putting additional argument "font_scale"***

***Check Below:-->>>>***

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

**In Below examples different contexts as "paper","talk" and "notebook" have been ploted**

***Check Below:--->>>***

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('talk')
sns.countplot(x='sex',data=tips)
```
