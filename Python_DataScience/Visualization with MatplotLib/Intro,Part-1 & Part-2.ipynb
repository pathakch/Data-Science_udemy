## Figure Size & DPI

**DPI=dot per inch Or pixels per inch**

*Check Below* --->>>>

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('X Vs Y')
ax.set_xlabel('X---->>')
ax.set_ylabel('Y=X^2')
```

In the above example  in **fig=plt.figure(figsize=(3,2))**

**3** is width of figure and **2** is height of figure in inch

In **ax=fig.add_axes([0,0,1,1])** 1,1 defines width of figure and height of figure

## Setting legends in plots 

In **ax.plot(x,x^2,label='X Squared')** label has been put as next argument 

and then with the help of **ax.legend(loc=0)** legends have been created

**(loc=0) defines the place of legend in picture** 

**There are many values for "loc" which define the different places of legends in figure**



```python
fig=plt.figure(figsize=(3,2))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.set_title('X Vs X**2 & X Vs X**3')
ax.set_xlabel('X--')
ax.set_ylabel('Y--')

ax.legend(loc=0)
```

**We can pass a tuples for many values in "loc" to decide the different places of legends in picture**

***Check Below Example***--->>

```python
fig=plt.figure(figsize=(5,3))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.set_title('X Vs X**2 & X Vs X**3')
ax.set_xlabel('X--')
ax.set_ylabel('Y--')

ax.legend(loc=(0.1,0.4))
```

## Linestyle in Plottig:

***Check Below***--->>>

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,linewidth=5,alpha=1,linestyle='--')
```

"***linestyle='steps'***  " OR " ***linestyle=':'***  "

Short form of *linestyle* is " ***ls*** " and *linewidth* is " ***lw***  "

## Adding *marker* and *markersize* in plots

***Check Below:--->>>>***

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Black',lw=2,alpha=2,ls='-',marker='+',markersize='20')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Black',lw=2,alpha=2,ls='-',marker='o',markersize='20'
        ,markeredgewidth=4,markeredgecolor='red',markerfacecolor='yellow')
```

## Setting ***x limit*** and ***y limit*** on Plots 

***Check Below***:--->>

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',lw=3,ls='--',marker='o',markeredgewidth='3'
        ,markeredgecolor='black',markerfacecolor='yellow',markersize=15)
ax.set_xlim([0,1])
ax.set_ylim([0,3])
```
